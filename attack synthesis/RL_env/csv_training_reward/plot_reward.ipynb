```python
import plotly.express as px
import pandas as pd
import plot_formatting
import plotly.graph_objects as go
```

```python
data = pd.read_csv("./run-TRPO_14-tag-rollout_ep_rew_mean.csv")
```

```python
data
```

```python
fig = px.line(data, x="Step", y="Value")
plot_formatting.beautify_plot(fig)
fig['layout'].update(width=800, height=250)
layout = go.Layout( autosize=True, margin={'l': 0, 'r': 0, 't': 0, 'b': 0})
fig['layout'].update(layout)
fig.update_layout(
    xaxis_title="Time",
    yaxis_title="Reward")
fig.show()
```

```python
fig.write_image('reward.pdf')
```
